```python
#importación librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def impseq(n0,n1,n2):
  #x(n)=delta(n-n0);n1<=n<=n2
  n=np.arange(n1,n2+1)
  x=(n-n0)==0
  return[x,n]
```

```python
#PUNTO 1
def unitario(n,n1, n2):

    u = np.where(n >= n1, 1, 0)  #  where es la función que genera secuencia del escalón unitario
    return(u)

n = np.arange(-6, 6+1)

n1 = 1
n2 = 5

y = unitario(n, n1, n2)

# Grafica la función escalon unitario como prueba

plt.plot(n, y,'m')
plt.stem(n, y,'m')
plt.title('punto1:Unitario')
plt.xlabel("rango")
plt.grid()
plt.show()
```

```python
#PUNTO 2
def ramp(n1,n2):
    n = np.arange(n1, n2 + 1)
    r = n - n1
    r[r < 0] = 0 #como en rampa no debe haber números negativos se pasa todos a 0
    return r, n
n1 = -6
n2 = 6

r, n = ramp(n1, n2)

plt.plot(n, r, 'c')
plt.stem(n, r, 'c')
plt.title('punto 2: Función de Rampa')
plt.xlabel('rango')
plt.grid()
plt.show()
```

```python
#PUNTO 2.b
def exp_complejas(n1, n2, a, e, f):
  n=np.arange(n1,n2+1)
  c=a*np.exp(e*n+f)
  return c, n

n1 = 0
n2 = 10
a = 1.0 #amplitud de la función compleja
e = 0.1 #es la base del exponente de la función exponencial compleja
f = 0.0 #constante que se suma al exponente de la función exponencial compleja

c, n = exp_complejas(n1, n2, a, e, f)


plt.plot(n, c,'aqua')
plt.stem(n, c,'aqua')
plt.xlabel('rango')
plt.title('punto2:Función exponencial compleja')
plt.grid()
plt.show()
```

```python
#PUNTO 3.a
def secuenciadelta_n():

    n = np.arange(-3, 2)#es este rango debido al teorema de delta, n solo puede tomar estos valores


    secuencia = 3 * (n == -1) + 5 * (n == -3) + 3 * (n == -2) + 3 * (n == 0) + (n == 1)

    return secuencia, n

secuencia, n = secuenciadelta_n()

plt.stem(n, secuencia,'gray')
plt.plot(n, secuencia,'gray')
plt.xlabel('rango')
plt.title('Punto 3(a)')
plt.grid()
plt.show()
```

```python
#PUNTO 3.b
def unitario(n1, n2, shift=0):
    n = np.arange(-6,6+1)# se inicializa en ese para poder graficar
    u = np.where(n >= n1, 1, 0)

    n -= shift

    return u, n

def main():

    u1,n1= unitario(0, -6)# especifico desde donde hasta donde se genera la función
    u2,n2= unitario(0, 6)

    v_1 = u1 * -2 # se genera una variable para poder multiplicarla por el correspondiente
    v_2 = u2 * 3

    resultado = v_1 + v_2

    n = n1

    return resultado, n

if __name__ == "__main__":
    resultado, n = main()#para poderse mostrar en el scrip
    plt.plot(n,resultado,'teal')
    plt.stem(n,resultado,'teal')
    plt.title('punto 3(b):secuencia')
    plt.xlabel('rango')
    plt.grid()
    plt.show()
```

```python
#PUNTO 3.c
def expo(n,n1, n2):
  e = np.exp(0.01*n )
  return e

def seno(n,n1, n2):
  s = np.sin(2 *np.pi*n )
  return s

n=np.arange(0,100+1)
n1 = 0
n2 = n[-1]

exponencial = expo(n,n1,n2)

sinusoidal = seno(n,n1,n2)

secuencia = exponencial * sinusoidal

plt.stem(n,secuencia,'green')
plt.plot(n,secuencia,'green')
plt.title('punto 3(c)')
plt.grid()
plt.show()
```

```python
#PUNTO 3.d

def unitario(n):

    u = np.where(n >= 0, 1, 0)
    return u

def rampa(n):

    r = np.maximum(n, 0)
    return r

n = np.arange(-10, 10+1)

u = unitario(n)#secuencia de escalón unitario para el rango de tiempo n.
r = rampa(n)#secuencia de rampa para el rango de tiempo n.

x4 = 2 * rampa(n + 3) - rampa(n - 2) - 5 * unitario(n - 3)#se pasa la secuencia mostrada en la guía


plt.stem(n, x4,'deeppink')

plt.xlabel('rango')
plt.title('punto 3(d)')
plt.grid()
plt.show()
```

```python
#PUNTO 4
def rampa(n):

    r = np.maximum(n, 0)
    return r


n = np.arange(0, 18)

ar = np.zeros_like(n)


ar[0:5+1] = rampa(n[0:5+1])
ar[6:11+1] = rampa(n[6:11+1] - 5)
ar[12:17+1] = rampa(n[12:17+1] - 10)

dar = np.diff(ar)

plt.subplot(2, 1, 1)
plt.stem(n, ar,'plum')
plt.title('Secuencia')
plt.xlabel('rango')

plt.grid()

plt.subplot(2, 1, 2)
plt.stem(n[1:], dar,'plum')
plt.title('Derivada')
plt.xlabel('rango')

plt.grid()

plt.tight_layout()
plt.show()
#En estas secuencias se puede ver rangos desplazados 5 en el eje x y se puede observar
#picos en donde hay cambios bruscos en la secuencia por este desplazamiento en la derivada del caso continuo no coincide completamente
#ya que la naturaleza de los datos es discreta.
```
